```python
# -*- coding: utf-8 -*-
"""
Семантический grep на Google Colab (Word2Vec russe)
Поиск строк по слову и его синонимам (леммы + POS)
Fr : Grep intelligent avec synonymes
"""

import os, sys
import gensim.downloader as api

# 1) загрузка модели
print("Загрузка модели Word2Vec...", file=sys.stderr)
model = api.load("word2vec-ruscorpora-300")   # лемматизированная модель
print("✅ Модель загружена")

# 2) параметры
topn      = 5
threshold = 0.70
file_name = "data.txt"

# 3) слово для поиска
word_raw = "привет"
target_words = {word_raw.lower()}   # точное слово обязательно

# 4) берём ВСЕ подходящие леммы и объединяем синонимы
keys = [k for k in model.key_to_index.keys() if word_raw in k]
if keys:
    print(f"Ключи для синонимов: {keys}", file=sys.stderr)
    for k in keys:
        try:
            for w, score in model.most_similar(k, topn=topn):
                if score >= threshold:
                    target_words.add(w.lower())
        except KeyError:
            continue
else:
    print(f"⚠ Ключи для «{word_raw}» не найдены -- ищем только точное совпадение", file=sys.stderr)

print(f"Ищем: {', '.join(sorted(target_words))}", file=sys.stderr)

# 5) поиск и вывод
with open(file_name, "r", encoding="utf-8") as f:
    for lineno, line in enumerate(f, 1):
        if any(w in line.lower() for w in target_words):
            print(f"{lineno}:{line}", end="")
```
